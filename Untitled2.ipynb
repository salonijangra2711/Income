# Data

The Adult dataset consists of the following columns:

Age: Represents the age of the individual.

Workclass: Describes the type of employment or work class, such as Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, State-gov, etc.

Education: Indicates the highest level of education achieved by the individual, such as Bachelors, Masters, Doctorate, etc.

Education-num: Represents the numerical representation of education, usually mapping to the number of years of education.

Marital-status: Describes the marital status of the individual, such as Married-civ-spouse, Divorced, Never-married, etc.

Occupation: Represents the occupation of the individual, such as Tech-support, Craft-repair, Sales, Exec-managerial, etc.

Relationship: Indicates the relationship status of the individual, such as Husband, Wife, Own-child, Unmarried, etc.

Race: Represents the race of the individual, such as White, Black, Asian-Pac-Islander, Amer-Indian-Eskimo, etc.

Sex: Indicates the gender of the individual, either Male or Female.

Capital-gain: Represents the capital gains of the individual.

Capital-loss: Represents the capital losses of the individual.

Hours-per-week: Indicates the number of working hours per week.

Native-country: Represents the country of origin of the individual.

Income: The target variable or label, indicating whether the individual's income is above or below $50,000 per year.

Please note that the specific dataset implementation or variations of the Adult dataset may have additional columns or slight 
differences in column names, but these are the common features present in the dataset.
